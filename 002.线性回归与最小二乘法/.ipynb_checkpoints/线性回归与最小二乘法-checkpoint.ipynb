```python
import pandas as pd

df = pd.read_excel('工龄-薪水.xlsx')

df.head()
```

```python
X = df[['工龄']]
Y = df['薪水']

%matplotlib inline
import matplotlib.pyplot as plt

# 使图表中的中文字符正常显示
plt.rcParams["font.family"]="Microsoft YaHei"

plt.scatter(X, Y)
plt.xlabel('工龄')
plt.ylabel('薪水')
plt.title('工龄-薪水')
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, Y)
r = model.predict(pd.DataFrame({'工龄':[10]}))
print(r, model.intercept_, model.coef_[0])

import numpy as np
Z = np.linspace(0, 10, 200)
plt.scatter(X, Y)
plt.plot(Z, model.predict(pd.DataFrame({'工龄':Z})))
plt.show()
```

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split

import numpy as np

dataset = load_diabetes()

X_train, X_test, y_train, y_test = train_test_split(dataset.data, dataset.target, random_state=0)

ones = np.ones((X_train.shape[0], 1))

X = np.hstack((ones, X_train))

theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y_train)


def predict(X_predict):
    ones = np.ones((X_predict.shape[0],1))
    X = np.hstack((ones, X_predict))
    return X.dot(theta)


print(theta)

def mean_squarred_error(y, y_predict):
    return np.sum((y - y_predict)**2) / len(y)

def r2_score(X_test, y_test):
    return 1 -  mean_squarred_error(y_test, predict(X_test)) / np.var(y_test)

score = r2_score(X_test, y_test)

print(score)

```
