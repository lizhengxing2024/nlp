```python
import pandas as pd

df = pd.read_excel('工龄-薪水.xlsx')

df.head()
```

```python
X = df[['工龄']]
Y = df['薪水']

%matplotlib inline
import matplotlib.pyplot as plt

# 使图表中的中文字符正常显示
plt.rcParams["font.family"]="Microsoft YaHei"

plt.scatter(X, Y)
plt.xlabel('工龄')
plt.ylabel('薪水')
plt.title('工龄-薪水')
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, Y)
r = model.predict(pd.DataFrame({'工龄':[4]}))
print(r)

import numpy as np
Z = np.linspace(0, 10, 200)
plt.scatter(X, Y)
plt.plot(Z, model.predict(pd.DataFrame({'工龄':Z})))
plt.show()
```
